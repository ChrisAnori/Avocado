```python
# 4.1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose   # para descomponer la serie temporal de precios en componentes de tendencia, estacionalidad y ruido.

# Cargar el conjunto de datos
data = pd.read_csv('/home/jose/avocado/Avocado/regresion/avocado.csv')
data_avo=data.copy() 

regiones_geograficas= ["California","GreatLakes","Midsouth","Northeast","Plains","SouthCentral","Southeast","West"]
data_avo=data_avo[data_avo['region'].isin(regiones_geograficas)]
data_avo['Date'] = pd.to_datetime(data_avo['Date'])


# Agrupar por trimestre y calcular el promedio de precios y la suma de volmenes
cohortes_trimestrales = data_avo.groupby(pd.Grouper(key='Date', freq='Q')).agg({
    'AveragePrice': 'mean',
    'Total Volume': 'sum'
}).reset_index()
print(cohortes_trimestrales.head(10) )  # First 10 rows
print(cohortes_trimestrales.info() )
# Crear el grfico
fig, ax1 = plt.subplots(figsize=(12, 6))

# Grfico de la evolucin de precios promedio en el eje y izquierdo
ax1.plot(cohortes_trimestrales['Date'], cohortes_trimestrales['AveragePrice'], color='blue', marker='o', label='Precio Promedio')
ax1.set_xlabel('Trimestres')
ax1.set_ylabel('Precio Promedio', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_title('Evolucin Trimestral de Precios Promedios y Volumen Total de Ventas')

# Crear el segundo eje y para el volumen total de ventas derecho 
ax2 = ax1.twinx()  # Crear un segundo eje y que comparte el eje x con el primero
ax2.plot(cohortes_trimestrales['Date'], cohortes_trimestrales['Total Volume'], color='green', marker='o', label='Volumen Total')
ax2.set_ylabel('Volumen Total', color='green')
ax2.tick_params(axis='y', labelcolor='green')


# Aadir leyendas y ajuste del layout
fig.tight_layout()  # Ajusta el layout para evitar superposicin
plt.show()
```

```python
# 4.2 
# Agrupar los datos por 'region' y 'Date', calculando el promedio de AveragePrice y la suma de Total Volume
cohort_region = data_avo.groupby(['region', pd.Grouper(key='Date', freq='QE')]).agg({
    'AveragePrice': 'mean',
    'Total Volume': 'sum'
}).reset_index()

# Crear una columna para ao y trimestre en formato 'YYYY-Tn' (T1, T2, T3, T4)
cohort_region['trimestre'] = cohort_region['Date'].dt.to_period('Q').astype(str)

# para la visualizacin
plt.figure(figsize=(14, 7))

# Grfico de barras para Average Price por regin
plt.subplot(2, 1, 1)
sns.barplot(data=cohort_region, x='trimestre', y='AveragePrice', hue='region', palette='Set1')
plt.title('Promedio de Precios por Regin y Trimestre')
plt.xlabel('Fecha (Ao-Trimestre)')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.legend(title='Region', loc='upper left', bbox_to_anchor=(1, 1))  # Adjust legend position
plt.tight_layout()


plt.figure(figsize=(14, 7))

# Grfico de barras para Average Price para Total Volume
plt.subplot(2, 1, 1)
sns.barplot(data=cohort_region, x='trimestre', y='Total Volume', hue='region', palette='Set1')
plt.title('Suma Total Volume por Regin y Trimestre')
plt.xlabel('Fecha (Ao-Trimestre)')
plt.ylabel('Total Volume')
plt.xticks(rotation=45)
plt.legend(title='Region', loc='upper left', bbox_to_anchor=(1, 1))  # Adjust legend position
plt.tight_layout()
```

```python
#  Utilizo el daset original para hacer una copia con las region California ( Santa Clara) y Houston 


''' 

En 2016 y 2017, los Super Bowl se llevaron a cabo en ciudades especficas:
	Super Bowl 50 (2016): Santa Clara, California. 7 de febrero de 2016. semana 5 
	Super Bowl LI (2017): Houston, Texas.  1 de febrero de 2017. semana 5 . region SouthCentral
Por lo tanto, en este anlisis, nos enfocaremos nicamente en los datos de las regiones relacionadas con estas ubicaciones: California y Houston.
filtraremos nicamente los datos de 2016 y 2017 para estas regiones ya que 2015 esta incompleto y lo comparamos con otro perodo en las mismas regiones/ciudades
'''
data_superbowl= data.copy() 
data_superbowl['Date'] = pd.to_datetime(data_superbowl['Date'])

# Crear una columna de semana para facilitar el grupo
data_superbowl['year'] = data_superbowl['Date'].dt.year
data_superbowl['week'] = data_superbowl['Date'].dt.isocalendar().week

# Agrupar por ao, semana y regin y calcular el promedio de precio y suma de volumen
grouped_data = data_superbowl.groupby(['year', 'week', 'region']).agg({
    'AveragePrice': 'mean',
    'Total Volume': 'sum'
}).reset_index()

# Filtrar los aos y las regiones especficas
#filtered_data = grouped_data[(grouped_data['year'].isin([2016, 2017])) &
#                             (grouped_data['region'].isin(['California', 'Houston']))]
filtered_data = grouped_data[(grouped_data['year'].isin([2016, 2017])) &
                             (grouped_data['region'].isin(['California', 'SouthCentral']))]

# Definir las semanas para Super Bowl y Abril
filtered_data['Periodo'] = filtered_data['week'].apply(lambda x: 'Super Bowl' if 4 <= x <= 5 else ('Enero' if 1 <= x <= 2 else None))
filtered_data = filtered_data.dropna(subset=['Periodo'])  # Filtrar solo las filas con valores en 'Periodo'

# Crear grficos de caja para comparar volmenes y precios
plt.figure(figsize=(14, 6))

# Grfico de caja para Total Volume
plt.subplot(1, 2, 1)
sns.boxplot(data=filtered_data, x='Periodo', y='Total Volume', hue='region', palette='pastel')
plt.title('Comparacin del Volumen Total en Super Bowl vs Enero (California y Houston)')
plt.xlabel('Periodo')
plt.ylabel('Volumen Total')
plt.legend(title='Regin', loc='upper right')

# Grfico de caja para Average Price
plt.subplot(1, 2, 2)
sns.boxplot(data=filtered_data, x='Periodo', y='AveragePrice', hue='region', palette='muted')
plt.title('Comparacin del Precio Promedio en Super Bowl vs Enero (California y Houston)')
plt.xlabel('Periodo')
plt.ylabel('Precio Promedio ($)')
plt.legend(title='Regin', loc='upper right')

plt.tight_layout()
plt.show()

# Grfico de barras para Average Price
plt.subplot(1, 2, 2)
sns.barplot(data=filtered_data, x='Periodo', y='AveragePrice', hue='region', palette='muted', ci=None)
plt.title('Comparacin del Precio Promedio en Super Bowl vs Enero (California y Houston)')
plt.xlabel('Periodo')
plt.ylabel('Precio Promedio ($)')
plt.legend(title='Regin', loc='upper right')

plt.tight_layout()
plt.show()

#----
# Grfico de barras para Average Price
plt.subplot(1, 2, 2)
sns.barplot(data=filtered_data, x='Periodo', y='Total Volume', hue='region', palette='muted', ci=None)
plt.title('Comparacin del Total Volume en Super Bowl vs Enero (California y Houston)')
plt.xlabel('Periodo')
plt.ylabel('Total Volume')
plt.legend(title='Regin', loc='upper right')

plt.tight_layout()
plt.show()

# Filtrar los datos para California y el ao 2017
data_california_2017 = data_superbowl[(data_superbowl['region'] == 'California') & (data_superbowl['year'] == 2017)]

# Agrupar los datos por mes y sumar el Total Volume en cada mes
data_california_2017['month'] = data_california_2017['Date'].dt.month
volume_by_month = data_california_2017.groupby('month')['Total Volume'].sum()

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(volume_by_month, bins=12, color='skyblue', edgecolor='black')
plt.title('Distribucin del Volumen Total de Aguacates por Mes en California (2016)')
plt.xlabel('Volumen Total por Mes')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()


#---


# Filtrar los datos para California y mayo de 2017
data_california_may_jan_2017 = data_superbowl[(data_superbowl['region'] == 'California') & 
                                              (data_superbowl['year'] == 2017) & 
                                              (data_superbowl['Date'].dt.month == 2)]

'''data_california_may_jan_2017 = data_superbowl[(data_superbowl['region'] == 'California') & 
                                              (data_superbowl['year'] == 2017) & 
                                              ((data_superbowl['Date'].dt.month == 5) | (data_superbowl['Date'].dt.month == 1))] '''
print(data_california_may_jan_2017['Date'],data_california_may_jan_2017['Total Volume'])

```
